```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# mse_data_synchronous
file_path = "mse_data_0219/mse_random.csv"
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"error：file {file_path} not found")
    exit()
```

<h2>MSE Convergence Over Simulation Cycles</h2>

```python
if 'Avg' in df.columns:  # debug
    x_col = 'Cycle'
    y_col = 'MSE'
else: 
    x_col = df.columns[0]
    y_col = df.columns[1]

# Make sure the data type is correct
df[x_col] = pd.to_numeric(df[x_col], errors='coerce')
df[y_col] = pd.to_numeric(df[y_col], errors='coerce')

plt.figure(figsize=(14, 6))

# MSE graph
plt.plot(df[x_col], df[y_col], 
         linewidth=1.5, 
         marker='o' if len(df) < 1000 else None,
         markersize=4,
         alpha=0.8,
         color='#FF2D2D',
         label='MSE')

plt.title(f'Mean-Square Error Convergence Trend (n={len(df)})', fontsize=14)
plt.xlabel('Simulation Cycle', fontsize=12)
plt.ylabel('MSE', fontsize=12)
plt.grid(True, alpha=0.4)
plt.legend()


max_cycle = df[x_col].max()
if max_cycle > 100000:
    plt.xticks(range(0, max_cycle + 1, 100000))
elif max_cycle > 10000:
    plt.xticks(range(0, max_cycle + 1, 10000))
elif max_cycle > 1000:
    plt.xticks(range(0, max_cycle + 1, 1000))
elif max_cycle > 100:
    plt.xticks(range(0, max_cycle + 1, 10))


min_y, max_y = df[y_col].min(), df[y_col].max()
y_interval = (max_y - min_y) / 10
plt.yticks(np.arange(min_y, max_y + y_interval, y_interval))

plt.tight_layout()

save_path = file_path.replace('.csv', '_MSE_NoRestriction_plot.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"figure saved to：{save_path}")


plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

if 'Avg' in df.columns:  # debug
    x_col = 'Cycle'
    y_col = 'MSE'
else: 
    x_col = df.columns[0]
    y_col = df.columns[1]

df[x_col] = pd.to_numeric(df[x_col], errors='coerce')
df[y_col] = pd.to_numeric(df[y_col], errors='coerce')

df = df.iloc[:200] 

df = df[(df[x_col] > 0) & (df[y_col] > 0)]

# polynomial fitting 
degree = 3 
poly_coeffs = np.polyfit(np.log(df[x_col]), np.log(df[y_col]), degree)
poly_func = np.poly1d(poly_coeffs)
df['Poly_Fit'] = np.exp(poly_func(np.log(df[x_col])))

# exponential fitting
def exp_func(x, a, b, c):
    return a * np.exp(-b * x) + c

params, _ = curve_fit(exp_func, df[x_col], df[y_col], p0=(1, 1e-6, 0), maxfev=2000)
a_fit, b_fit, c_fit = params
df['Exp_Fit'] = exp_func(df[x_col], *params)

plt.figure(figsize=(14, 6))

plt.plot(df[x_col], df[y_col], 
         linewidth=1.5, 
         marker='o' if len(df) < 1000 else None,  
         markersize=4,
         alpha=0.8,
         color='#FF2D2D',
         label='MSE')

# polynomial curve
plt.plot(df[x_col], df['Poly_Fit'], 
         linestyle='dashed', 
         linewidth=2, 
         color='grey', 
         label=f'Polynomial Fit (deg={degree})')

# exponential curve
plt.plot(df[x_col], df['Exp_Fit'], 
         linestyle='dotted', 
         linewidth=2, 
         color='green', 
         label='Exponential Fit')

plt.title(f'Mean-Square Error Convergence Trend', fontsize=14)
plt.xlabel('Simulation Cycle (log scale)', fontsize=12)
plt.ylabel('MSE (log scale)', fontsize=12)
plt.xscale('log')  # logx
plt.yscale('log')  # logy
plt.grid(True, which='both', linestyle='--', alpha=0.4)
plt.legend()

    
# poly. formula
poly_text = r"$y = " + " + ".join([f"{coef:.2f} x^{degree-i}" if degree-i > 1 
                                     else f"{coef:.2f} x" if degree-i == 1 
                                     else f"{coef:.2f}" 
                                     for i, coef in enumerate(poly_coeffs)]) \
            .replace("+ -", "- ") + "$"

plt.text(0.05, 0.25, poly_text, transform=plt.gca().transAxes, 
         fontsize=12, color='grey', bbox=dict(facecolor='white', alpha=0.6))


# exp. formula
exp_text = f"$y = {a_fit:.2e} \cdot e^{{-{b_fit:.2e} x}} + {c_fit:.2e}$"
plt.text(0.05, 0.15, exp_text, transform=plt.gca().transAxes, 
         fontsize=12, color='green', bbox=dict(facecolor='white', alpha=0.6))


plt.tight_layout()

save_path = file_path.replace('.csv', '_MSE_loglog_fit_plot.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"figure saved to：{save_path}")

plt.show()
```

<h2>Combine MSE & Discrepancy</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


if 'Avg' in df.columns:
    x_col = 'Cycle'
    mse_col = 'MSE'
    discrepancy_col = 'Discrepancy'
else:
    x_col = df.columns[0]
    mse_col = df.columns[1]
    discrepancy_col = df.columns[5]

for col in [x_col, mse_col, discrepancy_col]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df[(df[mse_col] > 0) & (df[discrepancy_col] > 0)]

plt.figure(figsize=(14, 6))

# MSE
plt.plot(df[x_col], df[mse_col],
         linewidth=1.5,
         marker='o' if len(df) < 1000 else None,
         markersize=4,
         alpha=0.8,
         color='#FF2D2D',
         label='MSE')

# Discrepancy
plt.plot(df[x_col], df[discrepancy_col],
         linewidth=1.5,
         marker='s' if len(df) < 1000 else None,
         markersize=4,
         alpha=0.8,
         color='#1f77b4',
         label='Discrepancy')

# Calculate the rate of change of the difference to find the point where the change is big
mse_diff = np.abs(np.gradient(np.log10(df[mse_col].values)))
disc_diff = np.abs(np.gradient(np.log10(df[discrepancy_col].values)))
threshold = 0.3 


significant_changes = df[x_col].values[(mse_diff > threshold) | (disc_diff > threshold)]

# Draw a vertical line
for xc in significant_changes:
    plt.axvline(x=xc, color='gray', linestyle='--', linewidth=1, alpha=0.5)

plt.title(f'MSE & Discrepancy Convergence Trend', fontsize=14)
plt.xlabel('Simulation Cycle (log scale)', fontsize=12)
plt.ylabel('Value (log scale)', fontsize=12)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which='both', linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()


save_path = file_path.replace('.csv', '_combined_loglog_plot.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"figure saved to：{save_path}")


plt.show()
```
